```python
import pandas as pd
import plotly.express as px

movie_data = pd.read_csv('movies.csv') # leer los datos
fig = px.histogram(movie_data, x="genres") # crear un histograma
fig.show() # crear gráfico histograma

fig_dispersion = px.scatter(movie_data, x="genres", y="popularity") # crear un gráfico de dispersión
fig_dispersion.show() # crear gráfico de dispersión
```
